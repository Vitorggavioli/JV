```python
import random
import time
from datetime import datetime

def monitor_temperatura(intervalo, limite_min, limite_max):
    while True:
        # Gera temperatura aleatória entre -198 e -192
        temperatura = round(random.uniform(-198, -192), 2)

        # Obtém data e hora atuais
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        dia = datetime.now().strftime("%A")  # Dia da semana

        # Mostra as informações no terminal
        print(f"[{timestamp} - {dia}] Temperatura: {temperatura}°C")

        # Verifica se a temperatura está fora do intervalo permitido
        if temperatura < limite_min or temperatura > limite_max:
            print(f"🚨 ALERTA! Temperatura fora do limite: {temperatura}°C 🚨")

            # Registra o alerta no arquivo
            with open("alertas_temperatura.txt", "a") as alerta:
                alerta.write(f"[{timestamp} - {dia}] ALERTA: {temperatura}°C fora do limite!\n")

            # Permite correção manual
            while True:
                try:
                    print("🔧 Corrija a temperatura manualmente:")
                    nova_temperatura = float(input("Digite a nova temperatura (-192 a -198): "))
                    if limite_min <= nova_temperatura <= limite_max:
                        temperatura = nova_temperatura
                        print(f"✅ Temperatura corrigida para {temperatura}°C.")
                        break
                    else:
                        print(f"⚠️ Valor inválido! Deve estar entre {limite_min} e {limite_max}.")
                except ValueError:
                    print("⚠️ Entrada inválida! Digite um número.")

        # Salva a temperatura no log
        with open("log_temperaturas.txt", "a") as log:
            log.write(f"[{timestamp} - {dia}] Temperatura: {temperatura}°C\n")

        # Aguarda pelo intervalo definido (em segundos)
        time.sleep(intervalo)

# Configuração do sistema
intervalo_em_segundos = 60  # Intervalo de 1 minuto
limite_minimo = -198        # Limite mínimo permitido
limite_maximo = -192        # Limite máximo permitido

# Inicia o monitoramento
monitor_temperatura(intervalo_em_segundos, limite_minimo, limite_maximo)
```

```python
import random
import time
from datetime import datetime

# Função que monitora a temperatura
def monitorar_temperatura():
    while True:
        # Gera temperatura aleatória, incluindo valores fora do limite para simular problemas
        temperatura = round(random.uniform(-200, -190), 2)

        # Obtém a data e hora atuais
        agora = datetime.now()
        data_hora = agora.strftime("%d/%m/%Y %H:%M:%S")
        dia_da_semana = agora.strftime("%A")  # Exemplo: Segunda-feira

        # Mostra a temperatura no terminal
        print(f"{dia_da_semana}, {data_hora} - Temperatura: {temperatura}°C")

        # Verifica se a temperatura está fora do limite
        if temperatura < -198 or temperatura > -192:
            print(f"🚨 ALERTA: Temperatura fora do limite permitido ({temperatura}°C).")
            print("Por favor, corrija a temperatura para que o sistema continue funcionando.")

            # Solicita ao usuário uma correção manual
            while True:
                try:
                    nova_temperatura = float(input("Digite uma nova temperatura válida (-192 a -198): "))
                    if -198 <= nova_temperatura <= -192:
                        temperatura = nova_temperatura
                        print(f"✅ Problema resolvido! Nova temperatura: {temperatura}°C.")
                        break
                    else:
                        print("⚠️ A temperatura deve estar entre -192 e -198. Tente novamente.")
                except ValueError:
                    print("⚠️ Entrada inválida! Por favor, digite um número.")

        # Registra a temperatura em um arquivo de log
        with open("log_temperatura.txt", "a") as log:
            log.write(f"{dia_da_semana}, {data_hora} - Temperatura: {temperatura}°C\n")

        # Aguarda 1 minuto antes da próxima leitura
        time.sleep(60)

# Inicia o programa
monitorar_temperatura()
```
