```python
import random
import time
from datetime import datetime

def monitor_temperatura(intervalo, limite_min, limite_max):
    while True:
        # Gera temperatura aleatÃ³ria entre -198 e -192
        temperatura = round(random.uniform(-198, -192), 2)

        # ObtÃ©m data e hora atuais
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        dia = datetime.now().strftime("%A")  # Dia da semana

        # Mostra as informaÃ§Ãµes no terminal
        print(f"[{timestamp} - {dia}] Temperatura: {temperatura}Â°C")

        # Verifica se a temperatura estÃ¡ fora do intervalo permitido
        if temperatura < limite_min or temperatura > limite_max:
            print(f"ðŸš¨ ALERTA! Temperatura fora do limite: {temperatura}Â°C ðŸš¨")

            # Registra o alerta no arquivo
            with open("alertas_temperatura.txt", "a") as alerta:
                alerta.write(f"[{timestamp} - {dia}] ALERTA: {temperatura}Â°C fora do limite!\n")

            # Permite correÃ§Ã£o manual
            while True:
                try:
                    print("ðŸ”§ Corrija a temperatura manualmente:")
                    nova_temperatura = float(input("Digite a nova temperatura (-192 a -198): "))
                    if limite_min <= nova_temperatura <= limite_max:
                        temperatura = nova_temperatura
                        print(f"âœ… Temperatura corrigida para {temperatura}Â°C.")
                        break
                    else:
                        print(f"âš ï¸ Valor invÃ¡lido! Deve estar entre {limite_min} e {limite_max}.")
                except ValueError:
                    print("âš ï¸ Entrada invÃ¡lida! Digite um nÃºmero.")

        # Salva a temperatura no log
        with open("log_temperaturas.txt", "a") as log:
            log.write(f"[{timestamp} - {dia}] Temperatura: {temperatura}Â°C\n")

        # Aguarda pelo intervalo definido (em segundos)
        time.sleep(intervalo)

# ConfiguraÃ§Ã£o do sistema
intervalo_em_segundos = 60  # Intervalo de 1 minuto
limite_minimo = -198        # Limite mÃ­nimo permitido
limite_maximo = -192        # Limite mÃ¡ximo permitido

# Inicia o monitoramento
monitor_temperatura(intervalo_em_segundos, limite_minimo, limite_maximo)
```

```python
import random
import time
from datetime import datetime

# FunÃ§Ã£o que monitora a temperatura
def monitorar_temperatura():
    while True:
        # Gera temperatura aleatÃ³ria, incluindo valores fora do limite para simular problemas
        temperatura = round(random.uniform(-200, -190), 2)

        # ObtÃ©m a data e hora atuais
        agora = datetime.now()
        data_hora = agora.strftime("%d/%m/%Y %H:%M:%S")
        dia_da_semana = agora.strftime("%A")  # Exemplo: Segunda-feira

        # Mostra a temperatura no terminal
        print(f"{dia_da_semana}, {data_hora} - Temperatura: {temperatura}Â°C")

        # Verifica se a temperatura estÃ¡ fora do limite
        if temperatura < -198 or temperatura > -192:
            print(f"ðŸš¨ ALERTA: Temperatura fora do limite permitido ({temperatura}Â°C).")
            print("Por favor, corrija a temperatura para que o sistema continue funcionando.")

            # Solicita ao usuÃ¡rio uma correÃ§Ã£o manual
            while True:
                try:
                    nova_temperatura = float(input("Digite uma nova temperatura vÃ¡lida (-192 a -198): "))
                    if -198 <= nova_temperatura <= -192:
                        temperatura = nova_temperatura
                        print(f"âœ… Problema resolvido! Nova temperatura: {temperatura}Â°C.")
                        break
                    else:
                        print("âš ï¸ A temperatura deve estar entre -192 e -198. Tente novamente.")
                except ValueError:
                    print("âš ï¸ Entrada invÃ¡lida! Por favor, digite um nÃºmero.")

        # Registra a temperatura em um arquivo de log
        with open("log_temperatura.txt", "a") as log:
            log.write(f"{dia_da_semana}, {data_hora} - Temperatura: {temperatura}Â°C\n")

        # Aguarda 1 minuto antes da prÃ³xima leitura
        time.sleep(60)

# Inicia o programa
monitorar_temperatura()
```
